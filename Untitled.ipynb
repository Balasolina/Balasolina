```python
import requests
import base64
import json

# Defina o nome de usuário do GitHub e o nome do repositório
github_username = 'Balasolina'
github_repository = 'https://github.com/Balasolina/dataset1.git'

# Defina o token de acesso pessoal do GitHub
github_token = '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'  # Certifique-se de que o token tenha permissão para fazer push no repositório

# Defina o caminho do arquivo ou diretório que deseja fazer upload
source_path = '/content/Vehicle_number_plate-2-2'  # Substitua 'seu_diretório' pelo caminho correto

# Defina o caminho do arquivo ou diretório no repositório do GitHub
github_path = 'https://github.com/Balasolina/dataset1.git'  # Substitua 'caminho/no/repositorio' pelo caminho desejado

# Leia o conteúdo do arquivo ou diretório
def read_file_contents(file_path):
    if os.path.isfile(file_path):
        with open(file_path, 'rb') as file:
            return file.read()
    return None

# Faça o upload do arquivo ou diretório para o repositório do GitHub
def upload_to_github(username, repository, token, path, content):
    url = f'https://api.github.com/repos/{username}/{repository}/contents/{path}'
    headers = {
        'Authorization': f'token {token}',
        'Content-Type': 'application/json',
    }
    data = {
        'message': 'Adicione arquivos',
        'content': base64.b64encode(content).decode('utf-8'),
    }
    response = requests.put(url, headers=headers, data=json.dumps(data))
    if response.status_code == 201:
        print(f'Upload bem-sucedido para {url}')
    else:
        print(f'Erro ao fazer upload para {url}. Status code: {response.status_code}')
        print(response.content)

# Leia o conteúdo do arquivo ou diretório
file_content = read_file_contents(source_path)

if file_content is not None:
    # Faça o upload do arquivo ou diretório para o repositório do GitHub
    upload_to_github(github_username, github_repository, github_token, github_path, file_content)
```
